```python
import numpy as np
from copy import deepcopy
```

```python
test = ['O....#....',
        'O.OO#....#',
        '.....##...',
        'OO.#O....O',
        '.O.....O#.',
        'O.#..O.#.#',
        '..O..#O..O',
        '.......O..',
        '#....###..',
        '#OO..#....']

data = np.genfromtxt('day14_input.txt', dtype=str, delimiter='\n', comments=None)
```

```python
def get_map(data):
    rock = {}
    roll = {}
    for i in range(0, len(data)):
        for j in range(0, len(data[i])):
            if data[i][j] == '#':
                rock[(i,j)] = '#'
            elif data[i][j] == 'O':
                roll[(i,j)] = 'O'
    return rock, roll

def roll_north(data, rock, roll):
    next_roll = {}
    for i in range(0, len(data)):
        for j in range(0, len(data[i])):
            if (i,j) in roll.keys() and i == 0:
                next_roll[(i,j)] = 'O'
            elif (i,j) in roll.keys():
                for k in range(i-1, -1, -1):
                    if (k,j) in rock.keys() or (k,j) in next_roll.keys():
                        next_roll[k+1,j] = 'O'
                        break
                    elif k == 0:
                        next_roll[k,j] = 'O'
                        break
    return next_roll

def score_roll(data, roll):
    score = 0
    for pos in roll.keys():
        score += abs(pos[0]-len(data))
    return score

def print_map(data, rock, roll):
    for i in range(0, len(data)):
        string = ''
        for j in range(0, len(data[i])):
            if (i,j) in rock.keys():
                string += rock[(i,j)]
            elif (i,j) in roll.keys():
                string += roll[(i,j)]
            else:
                string += '.'
        print(string)
    return
                

test_rock, test_roll = get_map(test)
#print_map(test, test_rock, test_roll)
#print()
test_roll2 = roll_north(test, test_rock, test_roll)
#print_map(test, test_rock, test_roll2)
#print()
print(score_roll(test, test_roll2))
data_rock, data_roll = get_map(data)
data_roll2 = roll_north(data, data_rock, data_roll)
print('Part 1 result:', score_roll(data, data_roll2))
```

```python
def roll_south(data, rock, roll):
    next_roll = {}
    for i in range(len(data)-1, -1, -1):
        for j in range(0, len(data[i])):
            if (i,j) in roll.keys() and i == len(data)-1:
                next_roll[(i,j)] = 'O'
            elif (i,j) in roll.keys():
                for k in range(i+1, len(data)):
                    if (k,j) in rock.keys() or (k,j) in next_roll.keys():
                        next_roll[k-1,j] = 'O'
                        break
                    elif k == len(data)-1:
                        next_roll[k,j] = 'O'
                        break
    return next_roll

def roll_west(data, rock, roll):
    next_roll = {}
    for j in range(0, len(data[0])):
        for i in range(0, len(data)):
            if (i,j) in roll.keys() and j == 0:
                next_roll[(i,j)] = 'O'
            elif (i,j) in roll.keys():
                for k in range(j-1, -1, -1):
                    if (i,k) in rock.keys() or (i,k) in next_roll.keys():
                        next_roll[i,k+1] = 'O'
                        break
                    elif k == 0:
                        next_roll[i,k] = 'O'
                        break
    return next_roll

def roll_east(data, rock, roll):
    next_roll = {}
    for j in range(len(data[0])-1, -1, -1):
        for i in range(0, len(data)):
            if (i,j) in roll.keys() and j == len(data[0])-1:
                next_roll[(i,j)] = 'O'
            elif (i,j) in roll.keys():
                for k in range(j+1, len(data[0])):
                    if (i,k) in rock.keys() or (i,k) in next_roll.keys():
                        next_roll[i,k-1] = 'O'
                        break
                    elif k == len(data[0])-1:
                        next_roll[i,k] = 'O'
                        break
    return next_roll

def cycle(data, rock, roll, n):
    roll = deepcopy(roll)
    previous_roll = []
    scores = []
    for i in range(0, n):
        roll = roll_north(data, rock, roll)
        roll = roll_west(data, rock, roll)
        roll = roll_south(data, rock, roll)
        roll = roll_east(data, rock, roll)
        scores.append(score_roll(data, roll))
        
        for j in range(0, len(previous_roll)):
            if np.array_equal(sorted(list(previous_roll[j].keys())), sorted(list(roll.keys()))):
                start = j
                length = i - j
                rep_size = n-start
                rep_in_loop = rep_size//length
                extra = rep_size%length
                idx = start+extra-1
                
                return score_roll(data, previous_roll[idx])
        previous_roll.append(deepcopy(roll))
    return None
        
print(cycle(test, test_rock, test_roll, 1000000000))
print('Part 2 result:', cycle(data, data_rock, data_roll, 1000000000))
```
